```python
def compute_stats(distances, ns, thres=0.1):
    rejected_rate, recall = 0, 0
    tp, fn = 0, 0
    for l, r in ns['reliable']:
        if l >= distances.shape[0] or r >= distances.shape[1]:
            continue
        if distances[l + 1, r + 1] > thres:
            tp += 1
        else:
            fn += 1
    res = {
        'recall': tp / (tp + fn),
        'reject_rate': ((distances > -1) & (distances < thres)).sum() / (distances > -1).sum()
    }
    return res
```

```python
stats = compute_stats(all_distances, ns, thres=0.05)
print(f"recall = {stats['recall']}, reject_rate = {stats['reject_rate']}")
```

```python
def compute_dsc_distance(dsc1, dsc2):
    h1 = dsc1.h / dsc1.h.sum()
    h2 = dsc2.h / dsc2.h.sum()
#     hist_intersection = ((h1 + h2) / 2 * (1 - np.abs(h1 - h2))).sum()
    hist_intersection = (np.minimum(h1, h2) / 2 * (1 - np.abs(h1 - h2) ** 2)).sum()

    
    return hist_intersection
```
