```python
import numpy as np
import matplotlib.pyplot as plt

from quantization import get_all_fragments_colors, generate_palette_from_colors
from skimage.color import lab2rgb
```

```python
data_dir = '../voronoi/example'
colors = get_all_fragments_colors(data_dir)
palette = generate_palette_from_colors(colors)
print(palette)
```

```python
def visualize_palette(colors, lab=True):
    img = np.zeros((100, 50 * colors.shape[0] // 2, 3))
    for i, color in enumerate(colors):
        img[(i % 2) * 50: (i % 2) * 50 + 50, (i // 2) * 50: (i // 2) * 50 + 50] = color
    if lab:
        img=lab2rgb(img)
    plt.imshow(img)
    return img
```

```python
p = visualize_palette(palette)
```
